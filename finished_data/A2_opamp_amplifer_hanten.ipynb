```python
Cr=1.18*0.00001
Cf=0.961*0.00000001
Rr=0.998*1000
Rf=10.04*1000
def theoryall(w):
    return 20*np.log(abs(1j*w*Cr*Rf/(1j*Cr*w*Rr+1)/(1j*w*Cf*Rf+1)))/np.log(10)
def theory1(w):
    return 20*np.log(abs(1/(1/Rf+1j*w*Cf)/(Rr+1/(1j*w*Cr))))/np.log(10)
def theory2(w):
    return 20*np.log(abs(1/(1/Rf+1j*w*Cf)/(1/1j/w/Cr)))/np.log(10)
def theory3(w):
    return -20*np.log(abs((Rr+1/(1j*w*Cr))/Rf))/np.log(10)
def theory4(w):
    return 20*np.log(abs(Rf*Cr*1j*w))/np.log(10)
plt.xscale("log")
plt.plot(diff1[0],diff1[1],'.',label="R_r,C_f",color="green")
plt.plot(diff2[0],diff2[1],'.',label="C_f",color="red")
plt.plot(diff3[0],diff3[1],'.',label="R_r",color="blue")
plt.plot(diff4[0],diff4[1],'.',label="--",color="yellow")
plt.plot(diff1[0],theoryall(diff1[0]),'--',label="theorical_curve R_r,C_f",color="green")
plt.plot(diff2[0],theory2(diff2[0]),'--',label="theorical_curve C_f",color="red")
plt.plot(diff3[0],theory3(diff3[0]),'--',label="theorical_curve R_r",color="blue")
plt.plot(diff4[0],theory4(diff4[0]),'--',label="theorical_curve --",color="yellow")
plt.plot(diff1[0],-20*np.log(diff1[0])/np.log(10)+120,'--',label='open-loop-charastaristic',color='black')
#plt.plot(diff2[0],diff2[2],'.',label="20dB",color="green")
plt.xlabel("frequency[Hz](log-sacle)")
plt.ylabel("gain[dB]")
plt.title('Differentiating circuit')
#plt.axhline(37,ls='-',label="40-3[dB]")
#plt.axhline(17,ls='-',label="20-3[dB]",color='red')
#plt.axvline(x=63095,ls='--',color='red')
#plt.axvline(x=8912,ls='--')
plt.legend()
file1="diffgain.png"
plt.savefig(file1)
plt.show()
```
