```python
def coodinate_sort(x,y,alist,blist):
    absalist=np.abs(alist-x)
    #求めたい座標に一番近い点の値を探す
    #一番近い点からインデックスを求め対となるデータを探す
    index=np.argsort(absalist)[0]
    print("index={0}absalist[0]={1}".format(index,absalist[0]))
    print("指定したx座標に一番近いデータ点x:{0},y{1}".format(alist[index],blist[index]))
    absblist=np.abs(blist-y)
    np.sort(absblist)
    index=np.argsort(absblist)[0]
    print("指定したy座標に一番近いデータ点x:{0},y{1}".format(alist[index],blist[index]))
coodinate_sort(1000,but_LPF[1][0]-3,but_LPF[0],but_LPF[1])
    
    
```
