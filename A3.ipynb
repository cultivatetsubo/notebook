```python
import math
import numpy as np
from pandas import Series,DataFrame
import pandas as pd
import scipy.optimize
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
from datetime import datetime as dt
from matplotlib.font_manager import FontProperties

fp = FontProperties(fname=r'C:\Windows\Fonts\HGRGE.TTC', size=14)
txt=np.loadtxt("A3.txt").T
fanout=txt[0]
time=txt[1]
print("{}{}".format(fanout,t))
parameter_initial = np.array([0.0, 0.0])
def func(n,t,t_0):
    return t_0+n*t
paramater_optimal, covariance = scipy.optimize.curve_fit(func,fanout,time, p0=parameter_initial)
print("paramater ={}".format(paramater_optimal))
y = func(fanout,paramater_optimal[0],paramater_optimal[1])
plt.plot(fanout,time, 'o')
plt.plot(fanout, y, '-')
plt.xlabel('ファンアウト数',fontproperties=fp)
plt.ylabel('立ち上がり遅延時間[ns]',fontproperties=fp)
plt.legend([u'実験値プロット', 'フィッティング曲線'], loc='upper left',prop=fp)
plt.savefig('A3_1.png',bbox_inches='tight',fontproperties=fp)
```
