```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a_1_V = pd.read_csv("a_1_voltage.csv")
a_1_I  = pd.read_csv("a_1_current.csv")
a_2_V = pd.read_csv("a_2_voltage.csv")
a_2_I  = pd.read_csv("a_2_current.csv")
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel("time[s]")
ax1.set_ylabel("Voltage[V]")
ax1.scatter(a_1_V.iloc[:,3],a_1_V.iloc[:,4], s=1, c="r", label="voltage")
ax2 = ax1.twinx()  # 2つのプロットを関連付ける
ax2.set_ylabel("Current[A]")
ax2.scatter(a_1_I.iloc[:,3],a_1_I.iloc[:,4],s=1, c="b", label="current")
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1+h2, l1+l2)
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel("time[s]")
ax1.set_ylabel("Voltage[V]")
ax1.scatter(a_2_V.iloc[:,3],a_2_V.iloc[:,4], s=1, c="r", label="voltage")
ax2 = ax1.twinx()  # 2つのプロットを関連付ける
ax2.set_ylabel("Current[A]")
ax2.scatter(a_2_I.iloc[:,3],a_2_I.iloc[:,4],s=1, c="b", label="current")
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1+h2, l1+l2,loc="upper left")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.scatter(a_2_I.iloc[:,4],a_2_V.iloc[:,4], s=1)
plt.xlabel("Current[A]")
plt.ylabel("Voltage[V]")
plt.show()
```
