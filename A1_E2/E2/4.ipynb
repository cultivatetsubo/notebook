```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df3 = pd.read_csv("4-4.csv")
```

```python
Ig = np.array(df3.iloc[:,0])
Vm = np.array(df3.iloc[:,1])
Im = np.array(df3.iloc[:,2])
Ifm = np.array(df3.iloc[:,3])
v = np.array(df3.iloc[:,4])
Vg  = np.array(df3.iloc[:,5])
Ifg = np.array(df3.iloc[:,6])
```

```python
Rag=Ram
Pc=(Ig+Ifg)**2*Rag
P0_ = (P -Pc-Vg*(Ig+Ifg) - Vm*Ifm + Ram*Im**2)/2
Pm = Vg*(Ig+Ifg) + P0.mean() +Pc
Pm_ = Vg*(Ig+Ifg) + P0_ +Pc
T =  0.972 * Pm / v
T_ = 0.972 * Pm_ / v
P=Vm*(Im+Ifm)
eta = Pm/P*100
eta_ = Pm_/P*100
```

```python
#plt.scatter(Ig, P0, label="P0")
plt.scatter(Ig, Pc, label="Pc")
plt.scatter(Ig, Pm, label="Pm")
plt.scatter(Ig, Pm, label="Pm_")
plt.scatter(Ig, P0_, label="P0_")
plt.axhline(P0.mean(), label="P0", c="r")
plt.xlabel("Ig[A]")
plt.ylabel("Power[W]")
plt.legend()
plt.show()
plt.scatter(Ig, T)
plt.xlabel("Ig[A]")
plt.ylabel("T[kgm]")
plt.show()
plt.scatter(Ig, eta)
plt.xlabel("Ig[A]")
plt.ylabel("η")
plt.show()
```
