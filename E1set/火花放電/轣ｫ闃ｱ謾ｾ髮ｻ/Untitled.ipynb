```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties
fp = FontProperties(fname=r'C:\Windows\Fonts\HGRGE.TTC', size=14)
```

```python
df = pd.read_csv("sphia.csv")
```

```python
d1 = np.array(df.iloc[:,0])
V1 = np.array(df.iloc[:,1])
d2 = np.array(df.iloc[:,2])
V2 = np.array(df.iloc[:,3])
d3 = np.array(df.iloc[:,4])
V3 = np.array(df.iloc[:,5])
```

```python
plt.scatter(d1, V1, label="DC+")
plt.scatter(d2, V2, label="DC-")
plt.scatter(d3, V3, label="AC")
plt.xlabel("ギャップ距離d[mm]",fontproperties=fp)
plt.ylabel("火花電圧[kV]",fontproperties=fp)
plt.title("球ー球電極の平等電界",fontproperties=fp)
plt.legend()
plt.show()
plt.savefig("sphia.png")
```
