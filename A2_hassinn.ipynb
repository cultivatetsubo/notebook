```python
import numpy as np
from pandas import Series,DataFrame
import pandas as pd
import scipy.optimize
import matplotlib.pylab as plt
import seaborn as sns
import cmath
import math
%matplotlib inline
osc3=np.loadtxt('tyoudohassin_bode.txt').T
osc1=np.loadtxt('hassin_bode_1kΩ.txt').T
osc5=np.loadtxt('hassin_bode_5kΩ.txt').T
```

```python
def A(omega,k):
    C=0.00000001
    R=10000
    return 10*k*omega*C*R*1.0j/(1-(omega*C*R)**2+omega*C*R*3.0j)
A1=A(osc1[0]*2*np.pi,0.333)
A2=A(osc3[0]*2*np.pi,0.5)
A3=A(osc5[0]*2*np.pi,0.1)
print(A(1000,0.33))
print(A(1000,0.5))
print(np.angle(A(1000,0.33)))
print(np.angle(A(1000,0.5)))
print(0.0925147/0.30529)
print(0.14017/0.4625735)

gain1=20*np.log(abs(A(osc1[0]*2*np.pi,0.1)))*math.log10(math.e)
gain2=20*np.log(abs(A(osc3[0]*2*np.pi,0.333)))*math.log10(math.e)
gain3=20*np.log(abs(A(osc5[0]*2*np.pi,0.5)))*math.log10(math.e)
pahse2=np.angle(A2,deg=True)
pahse3=np.angle(A3,deg=True)
pahse1=np.angle(A1,deg=True)
#pahse1=B(osc1[0]*2*np.pi,0.1)*360.0/2.0/np.pi
#pahse2=B(osc3[0]*2*np.pi,0.3333)*360.0/2.0/np.pi
#pahse3=B(osc5[0]*2*np.pi,0.5)*360.0/2.0/np.pi
```
